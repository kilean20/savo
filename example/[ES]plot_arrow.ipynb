```python
from ES import ES
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def obj(x):
    y = 1-np.mean(x**2)
    return y
```

```python
es = ES(obj, 
        x0 = 0.5*np.ones(2),
        max_dx = 0.1*np.ones(2),
        x_bounds = np.array([(0,1)]*2)
       )
```

```python
es.runES(100)
```

```python
x = np.array(es.history['x'])
# Create a figure and axis
fig, ax = plt.subplots()
ax.scatter(x[:, 0], x[:, 1], c=es.history['y'], cmap='viridis')
ax.quiver(x[:-1, 0], x[:-1, 1], 
          x[1:, 0] - x[:-1, 0], 
          x[1:, 1] - x[:-1, 1], 
          scale_units='xy', angles='xy', scale=1, color='black',alpha=0.5)
```
